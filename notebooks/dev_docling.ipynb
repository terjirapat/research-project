```python
import sys
from pathlib import Path

sys.path.append(str(Path().resolve().parents[0]))

datapaper_path = '../data/paper'
```

```python
from docling.document_converter import DocumentConverter
```

```python
file_name = 'A SYSTEMATIC REVIEW OF DEMAND FORECASTING MODELS FOR.pdf'
path = f'{datapaper_path}/{file_name}'

converter = DocumentConverter() 
doc = converter.convert(path).document
```

```python
import re

def parse_markdown_sections(md: str):
    sections = {}
    current_title = None
    buffer = []

    for line in md.split("\n"):
        match = re.match(r"^#{1,6}\s+(.*)", line)

        if match:
            if current_title:
                sections[current_title] = "\n".join(buffer).strip()

            current_title = match.group(1)
            buffer = []
        else:
            buffer.append(line)

    if current_title:
        sections[current_title] = "\n".join(buffer).strip()

    return sections
```

```python
abstract = None
introduction = None
related_work = None
methodology = None
dataset = None
experiments = None
results = None
discussion = None
limitations = None
conclusion = None
future_work = None
references = None
appendix = None
```

```python
SECTION_ALIASES = {
    "abstract": ["abstract", "summary"],
    
    "introduction": ["introduction", "background"],
    
    "related_work": [
        "related work",
        "literature review",
        "previous work"
    ],
    
    "methodology": [
        "method",
        "methods",
        "methodology",
        "approach",
        "proposed method",
        "model"
    ],
    
    "dataset": [
        "dataset",
        "data",
        "data description",
        "data collection"
    ],
    
    "experiments": [
        "experiments",
        "experimental setup",
        "evaluation setup"
    ],
    
    "results": [
        "results",
        "experimental results",
        "evaluation"
    ],
    
    "discussion": [
        "discussion",
        "analysis"
    ],
    
    "limitations": [
        "limitations",
        "limitation"
    ],
    
    "conclusion": [
        "conclusion",
        "conclusions",
        "summary and conclusion"
    ],
    
    "future_work": [
        "future work",
        "future directions"
    ],
    
    "references": [
        "references",
        "bibliography"
    ],
    
    "appendix": [
        "appendix",
        "supplementary material"
    ]
}
```

```python
sections = parse_markdown_sections(doc.export_to_markdown())

paper = {k: "" for k in SECTION_ALIASES}

for title, content in sections.items():
    
    title_lower = title.lower()
    
    for key, aliases in SECTION_ALIASES.items():
        if any(alias in title_lower for alias in aliases):
            paper[key] += content
```
