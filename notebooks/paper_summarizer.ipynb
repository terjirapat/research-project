```python
from src.prompt.paper_summarizer import PromptPaperSummarizer
from src.llms.bedrock import BedrockNova
from src.utils.loader import load_json

nova_llm = BedrockNova(model_id="us.amazon.nova-micro-v1:0")
```

```python
paper = load_json('../data/paper_extracted/Application_of_machine_learning_techniques_for_supply_chain_demand_forecasting.json')
paper
```

```python
def dict_to_text(d):
    sections = []
    for k, v in d.items():
        sections.append(f"### {k}\n{v}")
    return "\n\n".join(sections)
```

```python
text = paper['abstract']
text = dict_to_text(text)
context = [{"role": "user", "content": text}]

response = nova_llm.run(PromptPaperSummarizer.abstract(), context)

print(f"{response.input_tokens} input tokens")
print(f"{response.output_tokens} output tokens")
print(response.content)
```
